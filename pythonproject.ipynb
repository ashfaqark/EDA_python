```python
gc = sns.countplot(x = 'Gender', data = df) # to plot bars for the count of gender

for bars in gc.containers: # to get labels on bars
    gc.bar_label(bars)
```

```python
sales_gen =df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x= 'Gender', y= 'Amount', data = sales_gen)
```

```python
Age_group_order = ['0-17','18-25','26-35', '36-45', '46-50','51-55',  '55+']
ag = sns.countplot(x = 'Age Group', hue = 'Gender', data = df, order = Age_group_order)

for bars in ag.containers: # to get labels on bars
    ag.bar_label(bars)
```

```python
Age_group_order = ['0-17','18-25','26-35', '36-45', '46-50','51-55',  '55+']
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group' , y = 'Amount', data = sales_age, order = Age_group_order)
```

```python
state_order = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set({'figure.figsize':(15,6)}) # to adjust the size of the chart
sns.barplot(x = 'State' , y = 'Orders', data = state_order)
```

```python
state_amount = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set({'figure.figsize':(15,6)}) # to adjust the size of the chart
sns.barplot(x = 'State' , y = 'Amount', data = state_amount)
```

```python
ms = sns.countplot(x = 'Marital_Status', data = df) # to plot bars for the count of marital status
sns.set({'figure.figsize':(6,5)}) # to adjust the size of the chart

for bars in ms.containers: # to get labels on bars
    ms.bar_label(bars)
```

```python
sales_state= df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' ,ascending = False)
sales_state
sns.set({'figure.figsize':(8,5)}) # to adjust the size of the chart

sns.barplot(x= 'Marital_Status', y = 'Amount', hue = 'Gender', data = sales_state)
```

```python
occ = sns.countplot(x = 'Occupation', data = df)
sns.set({'figure.figsize':(16,5)}) # to adjust the size of the chart

for bars in occ.containers: # to get labels on bars
    occ.bar_label(bars)
```

```python
state_amount = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set({'figure.figsize':(15,6)}) # to adjust the size of the chart
sns.barplot(x = 'Occupation' , y = 'Amount', data = state_amount)
```
